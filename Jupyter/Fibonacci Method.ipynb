# В чем заключается метод ?

Метод заключается в сужении интервала неопределенности от начально длины интервала b – a до длины, 
которая меньше заданной точности. 
Рассматриваемый метод Фибоначчи является методом одномерного поиска, 
поэтому для определения минимума функции f(x) на некоем заданном отрезке [a, b] и заданной точностью.

## Математическое описание метода Фибоначчи

Для сужения интервала неопределенности используются числа Фибоначчи. Это ряд, состоящий из целых чисел. 
Их особенность заключается в том, что каждый следующее число является суммой двух предыдущих.

 ## Формула для поиска чисел Фибоначчи:
    
 <h3>F0 = 1, FK = FK – 1 + FK – 2, где K = 2, 3, …, N.<h3>  _________________________________________________________________________________________________


С помощью чисел Фибоначчи определяются точки интервала неопределенности, в пределах которого будут вычисляться значения функции. 
Для начала нужно определить значения в точках X1 и X2, расположенных на одинаковых расстояниях от концов интервала [a, b]. 
X1 и X2 находятся по следующим формулам:
    
   <h3>X1 = a + (b - a) * FN / FN + 2 | X2 = a + b – X1<h3>
_________________________________________________________________________________________________   

После нахождения точек сравниваем f(X1) и f(X2), если f(X1) < f(X2), то новым интервалом неопределенности будет [a1, b1], в ином случае будет[a1, b1] = [X1, b]. 
Данный процесс сужения будет повторяться пока на некотором шаге K не получится [aK, bK], который сужается с помощью симметричных точек:
    <h3>XK1 = aK + (b - a) * FN – K + 1 / FN + 2            
        XK2 = aK + (b - a) * FN – K + 2 / FN + 2<h3>
        
При K = N цикл останавливается, и длина интервала неопределенности становится равной: 
        
        bN – aN = 2 * (b - a) / FN+2

## Полученные точки совпадают и делят отрезок [aN, bN] пополам и эта точка берется в качестве приближенного значения экстремума функции.
